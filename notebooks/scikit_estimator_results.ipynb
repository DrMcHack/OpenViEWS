```python
n = 100
a = np.random.randint(0,2,n)
b = np.random.beta(0.5, 0.5, n)
c = np.random.wald(5, 10, n)
d = b*c
col_outcome = ['a']
cols_features = ['b', 'c', 'd']
```

```python
df = pd.DataFrame({'a' : a,
                 'b' : b,
                 'c' : c,
                 'd' : d})
y = df.loc[:,col_outcome]
y = df.a
X = df.loc[:,cols_features]
X_p = df.loc[slice(0,1), cols_features]
```

```python
rf_500 =  RandomForestClassifier(n_estimators = 5, n_jobs=1)
rf_500.fit(X,y)
```
